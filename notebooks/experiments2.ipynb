```python
import imageio
import os

IMG_DIR = "/content/drive/MyDrive/Fall 2025/CMSC 498K/498K-Final-Project/externals/gaussian-splatting/output/hotdog-test/test/ours_30000/renders"
VIDEO_OUT = "/content/drive/MyDrive/Fall 2025/CMSC 498K/498K-Final-Project/externals/gaussian-splatting/output/hotdog-test/hotdog_render.mp4"

images = []
files = sorted(os.listdir(IMG_DIR))

for f in files:
    if f.endswith(".png"):
        images.append(imageio.imread(os.path.join(IMG_DIR, f)))

imageio.mimsave(VIDEO_OUT, images, fps=30)
print("Saved to:", VIDEO_OUT)
```

```python
from IPython.display import HTML
from base64 import b64encode

with open(VIDEO_OUT, 'rb') as video_file:
  video_bytes = video_file.read()
  video_base64 = b64encode(video_bytes).decode()
  html_player = f"""
    <video width="600" controls>
        <source src="data:video/mp4;base64,{video_base64}" type="video/mp4">
        Your browser does not support the video tag.
    </video>
    """

display(HTML(html_player))
```
